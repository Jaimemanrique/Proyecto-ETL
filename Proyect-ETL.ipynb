```python
url_fifa_22 = 'https://www.futwiz.com/en/fifa22/players?page=0&teams[]=243&order=rating&s=desc&release=nifgold'
```

```python
PATH=ChromeDriverManager().install()
driver=webdriver.Chrome(PATH)

driver.get(url_fifa_22)
```

```python
filas = driver.find_element(By.TAG_NAME, 'tbody')

data = [[e.text for e in f.find_elements(By.TAG_NAME, 'td')] 
      for f in filas.find_elements(By.TAG_NAME, 'tr')]
```
